```python
import pandas as pd

def fun(cell):
    if cell=='n.a.':
     return 'Sam Walton'
    return (cell)

def fun1(cell):
    if cell=='not available':
     return None
    return (cell)

df1=pd.read_excel('Book1.xlsx', index_col=0, converters= {
    'people': fun,
    'eps': fun1
})
df1
```

```python
df1.to_excel('Book2.xlsx', sheet_name='Stocks', startrow=1, startcol=2)
```

```python
import pandas as pd

table2020= pd.DataFrame({
    'cars':['ford','toyato','hyundai','kia','MG'],
    'rating':[3,2,5,1,1],
    'price':[1000000,1500000,600000,2000000,2500000]
})

table2021= pd.DataFrame({
    'cars':['ford','toyato','hyundai','kia','MG'],
    'rating':[3,2,5,1,1],
    'price':[1500000,1240000,680000,2500000,2800000]
})


with pd.ExcelWriter('Book3.xlsx') as writer:
    table2020.to_excel(writer, sheet_name='S2020')
    table2021.to_excel(writer, sheet_name='S2021')
```
