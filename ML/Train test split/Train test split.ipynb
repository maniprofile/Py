```python
x=df.drop('Sell Price($)',axis ='columns')
x
```

```python
x.dropna(how='all',axis=1,inplace=True)
x
```

```python
y=df['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=10)
```
