```python
import pandas as pd
df=pd.read_excel('Income.xlsx')
df
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Age')
plt.ylabel('Income')
plt.scatter(df.Age,df['Income($)'])
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
x=km.cluster_centers_
x
```

```python
df0=df[df['cluster']==0]
df1=df[df['cluster']==1]
df2=df[df['cluster']==2]

plt.xlabel('Age')
plt.ylabel('Income')

plt.scatter(df0.Age,df0['Income($)'],color='red',label='cluster0')
plt.scatter(df1.Age,df1['Income($)'],color='blue',label='cluster1')
plt.scatter(df2.Age,df2['Income($)'],color='green',label='cluster2')

plt.scatter(x[:,0],x[:,1],color='orange',marker='*',label='centroid')

plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
```

```python
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
```

```python
cluster=km.fit_predict(df[['Age','Income($)']])
df['cluster']=cluster
```

```python
df
```

```python
y=km.cluster_centers_
y
```

```python
df
```

```python
df0=df[df['cluster']==0]
df1=df[df['cluster']==1]
df2=df[df['cluster']==2]

plt.xlabel('Age')
plt.ylabel('Income')

plt.scatter(df0.Age,df0['Income($)'],color='red',label='cluster0')
plt.scatter(df1.Age,df1['Income($)'],color='blue',label='cluster1')
plt.scatter(df2.Age,df2['Income($)'],color='green',label='cluster2')

plt.scatter(y[:,0],y[:,1],color='orange',marker='*',label='centroid')

plt.legend()
plt.show()
```

```python
#Elbow plot



```

```python
sse=[]
k_range=range(1,10)
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python

plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_range,sse)
```

```python
df=pd.read_csv('Iris.csv')
df
```

```python
df=df.drop(['SepalLengthCm','SepalWidthCm'],axis=1)
df
```

```python
df=df.set_index('Id')
df
```

```python
df[:3]
```

```python
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.scatter(df.PetalLengthCm,df.PetalWidthCm)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
a=km.fit_predict(df[['PetalLengthCm','PetalWidthCm']])
a
```

```python
df['clusters']=a
df
```

```python
b=km.cluster_centers_
b
```

```python
df0=df[df['clusters']==0]
df1=df[df['clusters']==1]
df2=df[df['clusters']==2]

plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')

plt.scatter(df0.PetalLengthCm,df0['PetalWidthCm'],color='red',label='cluster0')
plt.scatter(df1.PetalLengthCm,df1['PetalWidthCm'],color='blue',label='cluster1')
plt.scatter(df2.PetalLengthCm,df2['PetalWidthCm'],color='green',label='cluster2')

plt.scatter(b[:,0],b[:,1],color='orange',marker='*',label='centroid')

plt.legend()
plt.show()
```

```python
#Elbow plot


```

```python
sse=[]
k_range=range(1,10)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['PetalLengthCm','PetalWidthCm']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_range,sse)
```
